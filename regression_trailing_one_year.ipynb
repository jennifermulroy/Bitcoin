```python
import pandas as pd
import numpy as np
import datetime 
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.express as px
import hvplot.pandas
import seaborn as sns
%matplotlib inline
from yahoofinancials import YahooFinancials
```

```python
#Trailing one year data 
start_sp = datetime.datetime(2018, 10, 1)
end_sp = datetime.datetime(2019, 10, 1)

#create dataframes

#Bitcoin 

BTC_df_year = pdr.get_data_yahoo('BTC-USD', start_sp, end_sp)

#SPDR S&P 500 ETF
SPX_df_year = pdr.get_data_yahoo('SPY', start_sp, end_sp)

#iShares Core U.S. Aggregate Bond ETF
AGG_df_year = pdr.get_data_yahoo('AGG', start_sp, end_sp)

#US Dollar/USDX - Index
USD_df_year = pdr.get_data_yahoo('DX-Y.NYB', start_sp, end_sp)

#Bloomberg Commodity Index Total Return ETN (DJCI)
DJCI_df_year = pdr.get_data_yahoo('DJCI', start_sp, end_sp)

#SPDR Gold Shares (GLD)
GLD_df_year = pdr.get_data_yahoo('GLD', start_sp, end_sp)
```

```python
#Drop extra columns to be left with closing price 
BTC_df_year.drop(columns=['High', 'Low', 'Open', 'Volume', 'Adj Close'], inplace=True)
BTC_df_year.head()
SPX_df_year.drop(columns=['High', 'Low', 'Open', 'Volume', 'Adj Close'], inplace=True)
SPX_df_year.head()
AGG_df_year.drop(columns=['High', 'Low', 'Open', 'Volume', 'Adj Close'], inplace=True)
AGG_df_year.head()
USD_df_year.drop(columns=['High', 'Low', 'Open', 'Volume', 'Adj Close'], inplace=True)
USD_df_year.head()
DJCI_df_year.drop(columns=['High', 'Low', 'Open', 'Volume', 'Adj Close'], inplace=True)
DJCI_df_year.head()
GLD_df_year.drop(columns=['High', 'Low', 'Open', 'Volume', 'Adj Close'], inplace=True)
```

```python
#rename closing column
BTC_df_year = BTC_df_year.rename(columns = {"Close": "BTC Daily Returns"}) 
SPX_df_year = SPX_df_year.rename(columns = {"Close": "SPX Daily Returns"}) 
AGG_df_year = AGG_df_year.rename(columns = {"Close": "AGG Daily Returns"}) 
USD_df_year = USD_df_year.rename(columns = {"Close": "USD Daily Returns"}) 
DJCI_df_year = DJCI_df_year.rename(columns = {"Close": "DJCI Daily Returns"}) 
GLD_df_year = GLD_df_year.rename(columns = {"Close": "GLD Daily Returns"}) 
```

```python
combined_df_year_price = pd.concat([BTC_df_year, SPX_df_year, AGG_df_year, USD_df_year, DJCI_df_year, GLD_df_year],axis='columns', join="inner")
combined_daily_return_year = combined_df_year_price.pct_change()
combined_daily_return_year = combined_daily_return_year.fillna(0, inplace=False)
combined_daily_return_year.head()
```

```python
correlation = combined_daily_return_year.corr()
correlation
```

```python
# Showcase beta vs. correlation by plotting a scatterplot using the Seaborn library and fitting a regression line

sns.lmplot(x='BTC Daily Returns', y='SPX Daily Returns', data= combined_daily_return_year, aspect=1.5, fit_reg=True)
```

```python
# Calculate Sharpe Ratio
sharpe_ratios = (combined_daily_return_year.mean() * 252) / (combined_daily_return_year.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Plot sharpe ratios
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```
